```python
%% sql

# Write your MySQL query statement below
-- Change name to warehouse_name
-- Join Product ID
-- Find volume of every product by product id
    -- volume is width * length * height
-- Multiply volume by number of units
-- Return results

SELECT name as warehouse_name
, SUM(units * (Width * Length * Height)) AS volume
FROM Warehouse a
JOIN Products b
ON a.product_id = b.product_id
GROUP BY warehouse_name
```

```python
%% sql

# Write your MySQL query statement below
-- find each month
-- find each country
-- sum the number of transactions 
-- sum the total amount of transactions
-- return results


SELECT country
, DATE_FORMAT(trans_date, '%Y-%m') AS month
, COUNT(state) AS trans_count
, SUM(CASE WHEN state = "approved" THEN 1 else 0 end) AS approved_count
, SUM(amount) as trans_total_amount
, SUM(CASE WHEN state = "approved" THEN amount else 0 end) as approved_total_amount
FROM Transactions
GROUP BY month, country
```

```python
%% sql

# Write your MySQL query statement below
-- Find percentage of immediate orders and round to two decimals
-- only return immediate+percentage column



SELECT 
    ROUND(SUM(
        IF(
            order_date = customer_pref_delivery_date, 1, 0)
    )/
          COUNT(delivery_id)*100,2) AS immediate_percentage 
FROM Delivery
```

```python
%% sql

# Write your MySQL query statement below
-- change from_id to person1 and to_id to person 2
-- count the number of times person 1 calls person 2
-- sum the call duration of when person 1 and person 2 have called each other 
-- filter to show number of times person 1 call duration is less than person 2
-- return results

SELECT LEAST(from_id,to_id) as person1
,GREATEST(from_id,to_id) as person2
,COUNT(*) as call_count
,SUM(duration) as total_duration
FROM Calls
GROUP BY person1
,person2;
```

```python
%% sql 

# Write your MySQL query statement below
-- Select sale_date
-- Case When fruit is apples else subtract and name diff
-- Sum this outcome
-- Return results


SELECT sale_date
, SUM(CASE WHEN fruit = 'apples' THEN sold_num else -sold_num end) as diff
FROM Sales
GROUP BY sale_date
```
